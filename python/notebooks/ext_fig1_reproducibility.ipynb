# Extended Data Figure 1 — Dimension reproducibility

Reproducibility of the 49 dimensions across 20 independent model fits.

```python
# Make the things_spose package importable whether this runs from python/ or
# python/notebooks/ (jupyter lab vs papermill/nbconvert set different cwds).
import sys, pathlib
for _c in [pathlib.Path.cwd(), *pathlib.Path.cwd().parents]:
    if (_c / "things_spose").is_dir():
        sys.path.insert(0, str(_c)); break

import matplotlib.pyplot as plt
from things_spose import dataio, backend
print(backend.describe())
ds = dataio.load_dataset()
SAVE = False  # set True to export SVGs next to this notebook
```

```python
from things_spose import reproducibility, viz
sortind0 = dataio.load_sortind()
res = reproducibility.dimension_reproducibility(ds.embedding, sortind0)
print(f'rank r = {res.rank_corr:.3f}, p = {res.rank_p:.4g}')
fig = viz.extfig1_reproducibility(ds, sortind0, result=res,
        save_path='ext_fig1.svg' if SAVE else None)
plt.show()
```
